```
# Importing the modules

import numpy as np  # To reshape data
import pandas as pd  # To convert the raw data into a dataframe
from ggplot import *  # For graphical visualizations
from sklearn.model_selection import train_test_split  # To split our data into training and testing data
from sklearn.linear_model import LinearRegression  # Classifier for our prediction algorithm
from sklearn import metrics  # To gauge the perfromance of our model
```

```
# Loading our data

url = "http://bit.ly/w-data"
Data = pd.read_csv(url)
print("Data Imported Successfully")
Data.head()
```

```
# Seperating the target and feature values

x = Data.iloc[:, :-1].values
y = Data.iloc[:, 1].values
```

```
# Visuallizing the data point and trying to pick a realtionship between data point of our data

ggplot(Data, aes(x='Hours', y='Scores')) + \
geom_point(x=x, y=y, color='steelblue', size=50) + \
ggtitle("Input Hours vs Score") + \
labs(x="Study Hours", y='Student Score')
```

```
#Splitting the raw data into training and testing set

X_train, X_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.2, random_state=0)

# Fitting the data on our model

regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

```
# Finding the model accuracy using the testing data

y_pred = regressor.predict(X_test)

# Displaying the results

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df
```

```
# Tesing with sample data

hours = 9.25
own_pred = regressor.predict(np.array(hours).reshape(1,-1))

print("No of Hours = ", hours)
print("Predicted Score = ", own_pred[0])
```

```
# Checking the model performance with the accuracy and mean absolute error values

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

print("Accuracy of the model is: ", regressor.score(X_test, y_test)*100,"%")
```
