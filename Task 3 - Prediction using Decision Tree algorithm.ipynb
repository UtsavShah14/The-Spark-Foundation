```
# Importing all the dependeneies

import sklearn.datasets  # For Iris Dataset
import pandas as pd  # To convert the raw data into a dataframe
import numpy as np  # To reshape data
from sklearn.tree import DecisionTreeClassifier  # Classifier for our prediction algorithm
from sklearn.externals.six import StringIO   # Make an object as file
from IPython.display import Image  # To display an Image from file
from sklearn.tree import export_graphviz  # Graphical representation of decision tree
import pydotplus  # For graphviz dot langugae representaion of graph
from sklearn.model_selection import train_test_split  # To split our data into training and testing data
```

```
# Loading the Iris dataset for our project
iris = sklearn.datasets.load_iris()

# Converting it into a dataframe
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)

# Viewing the data head
print(iris_df.head(5))

# Loading the target values for out decision tree classifier
y = iris.target
print("Target values: \n", y)
```

```
# Splitting the data into training and testing data

X_train, x_test, y_train, y_test = train_test_split(iris_df, y)
```

```
# The decision tree classifier with parameters
decision_tree_classifier = DecisionTreeClassifier(
    criterion = 'gini',
    splitter = 'best',
    max_depth = None,
    max_features = None,
    ccp_alpha = 0.0

)
# Fitting our data
decision_tree_classifier.fit(X_train, y_train)

# Depth of the Tree
print("The depth of the trained decision tree is:",decision_tree_classifier.get_depth())
```

```
# Testing the trained model
print("The accuraccy of our model is:",round(decision_tree_classifier.score(x_test, y_test)*100,2),"%")

# Predicting a new random sample with our model.
sample = [50, 30, 25, 15]
print("The output class for predicted class is:", decision_tree_classifier.predict(np.array(sample).reshape(1,-1)))
```

```
# Initializing the visuals for decision tree
dot_data = StringIO()
export_graphviz(
    decision_tree_classifier,
    out_file=dot_data,
    feature_names=iris.feature_names,
    filled=True
    )
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

# Converting and displaying it as an image
Image(graph.create_png())
```
